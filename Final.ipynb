## Ch-oo-p-pee / Ch-h-oo-p-pee [Silence / Hidden]
Ch-oo-p-pee: silence 
Ch-h-oo-p-pee: hidden 

The following is a series of explorations in silence leading to the final artifact that creates computer generated poetry using markov chains. 

The corpora of text I am using throughout these explorations are:  
1. Scattered journal entries from 2016-2019 (hand-typed from journal entries in March 2022. Two versions - one with date, one without and combined.) 
2. List of phrases used when describing a "silent" situation
3. List of quotes talking about silence from a scattering of readings done in prep for research in thesis 
4. List of words that relate to silence 
5. The Hindi alphabet in the devanagri script and its pronounciations in arpabet 
6. Transcribed conversations with people about meanings of silence and how silence manifests in their lives. 

That is a lot of lists and pieces of text to work with. There has to be a hidden story in these stories that I can find in this interrogation of silence. What is that going to be?

#### Exploration 2:  Replacing words identified in the silence word list with a space Using different words for silence in the quotes 
Visually replacing the word for silence with a blank or negative space. How does the meaning change when you change the word 'silence'?

```python
import tracery
```

```python
import string
regular_punct = list(string.punctuation) # python punctuations 
special_punct=['©', '^', '®',' ','¾', '¡','!'] # user defined special characters to remove 
def remove_punctuation(text,punct_list):
    for punc in punct_list:
        if punc in text:
            text = text.replace(punc, ' ')
    return text.strip()
```

First removing everything **except** the words for silence 

```python
statement = remove_punctuation(random.choice(people), regular_punct).lower().split()
print('original: \n' + ' '.join(statement))
print("\n")
new_statement = ''
for word in statement: 
    if word in s_words: 
        new_statement += word
    else: 
        for i in range(len(word)):
            new_statement += ' '
print('new: \n' + new_statement)
```

Now removing all the words for silence and leaving everything else.

### Final exploration

Combining the act of replacing words with spaces and markov chains. Taking the journal and conversations, breaking each into separate list of words. Taking the words of the conversation out of the journal, taking the words of the journal out of the conversation. Creating two new pieces of text. Creating a markov generator for each text. Combining those generators to create a final piece of text

Step 1: Reading the journal and the conversations and removing punctuation using the method defined above

```python
journal = remove_punctuation(open('combined_noDate.txt').read() + open('journal_hiding.txt').read(), regular_punct)
conversations = remove_punctuation(open('people_meanings.txt').read() + open('people_hiding.txt').read(), regular_punct)
```

```python
with open('final_journal', 'a') as f: 
    f.write(journal)
    
with open('final_conversations', 'a') as g: 
    g.write(conversations)
```

```python
print(journal)
```

```python
print(conversations)
```

Step 2: Splitting the two separate pieces of text into a list of words

```python
journal_words = journal.split()
conversations_words = conversations.split()
```

Step 3.1: Removing the words of the conversation from the journal 

```python
new_journalMinusConversations = ''
for i in range(len(journal_words) - 1):
    word = journal_words[i]
    if word in conversations_words: 
        for i in range(len(word)):
            new_journalMinusConversations += ' '
    else: 
        new_journalMinusConversations += ' ' + word + ' '
print(new_journalMinusConversations)
```

Step 3.2: Removing the words of the journal from the conversations

```python
new_conversationsMinusJournal = ''
for i in range(len(conversations_words) - 1):
    word = conversations_words[i]
    if word in journal_words: 
        for i in range(len(word)):
            new_conversationsMinusJournal += ' '
    else: 
        new_conversationsMinusJournal += ' ' + word + ' '
print(new_conversationsMinusJournal)
```

Step 3.3: Writing two new .txt files with the new pieces of redacted text

```python
with open('final_journalMinusConversations', 'a') as f: 
    f.write(str(new_journalMinusConversations))
    
with open('final_conversationsMinusJournal', 'a') as g: 
    g.write(str(new_conversationsMinusJournal))
```

Step 4: creating a markov generator with each of the two new texts

```python
gen_journalMinusConversations = markovify.Text(str(new_journalMinusConversations))
gen_conversationsMinusJournal = markovify.Text(str(new_conversationsMinusJournal))
```

```python
print(gen_journalMinusConversations.make_sentence())
print(gen_journalMinusConversations.make_short_sentence(100))
```

```python
print(gen_conversationsMinusJournal.make_sentence())
print(gen_conversationsMinusJournal.make_short_sentence(100))
```

Step 5: Combining both generators. 

Higher emphasis on the journal for artistic reasons.  We place higher emphassis on our mind than we do in conversations. 


```python
gen_combo_minus = markovify.combine([gen_journalMinusConversations, gen_conversationsMinusJournal], [2,0.5])
```

```python
print(gen_combo_minus.make_sentence())
print(gen_combo_minus.make_short_sentence(100))
```

Adding catch for 'None' exception 

```python
list_gen_minus = []
for i in range(50):
    a = gen_combo_minus.make_sentence()
    if a is not None:
        list_gen_minus.append(a)
        
print(list_gen_minus)
```

##### Final: making a list of 1000 generated sentences and adding it to a new text file to use elsewhere

```python
list_gen_minus = []
for i in range(1000):
    a = gen_combo_minus.make_sentence()
    if a is not None:
        list_gen_minus.append(a)

print(list_gen_minus)
with open('final_markovSentences', 'a') as f: 
    f.write('\n'.join(list_gen_minus))
```

Step 6: final formatting of poems using tracery grammar

```python
import tracery
```

```python
from tracery.modifiers import base_english
```

```python
markovSentences = open('final_markovSentences').read().split('\n')
markovSentences_long = []
markovSentences_short = []
for i in range(len(markovSentences) - 1): 
    s = markovSentences[i-1]
    if len(s) >= 100: 
        markovSentences_long.append(s)
    elif len(s) < 100 and len(s) >= 25: 
        markovSentences_short.append(s)
```

```python
rules = {
    "origin": "#long.capitalize# \n #short.capitalize# \n \n #short.capitalize# \n #long.capitalize#",
    "long": markovSentences_long,
    "short": markovSentences_short
}
grammar = tracery.Grammar(rules)
grammar.add_modifiers(base_english)
s = grammar.flatten("#origin#")
print(s)
```

```python
for i in range(100):
    s = grammar.flatten("#origin#")
    with open('final_final_poems', 'a') as final: 
        final.write('#' + str(i) + ':')
        final.write('\n')
        final.write(s)
        final.write('\n')
        final.write('\n')
```
